```python
import pandas as pd
```

```python
#Cash

csv_file_path_cash=r'C:\Users\Admin\OneDrive - VENDIMAN PVT LTD\Amanda\DSR\vendekin\cash 4 feb.csv'
df_cash = pd.read_csv(csv_file_path_cash)
df_cash['PaymentStatus']= 'SUCCESS'
df_cash['Type'] = 'Cash'
cash = df_cash.loc[:, ('Qrcode','Date', 'ProductName', 'TotalPrice', 'Location', 'Quantity','PaymentStatus', 'Type')]
cash.rename(columns={'TotalPrice': 'Price'}, inplace=True)
cash.head()
#df_cash.head()
```

```python
#Cashless

csv_file_path_cashless =r'C:\Users\Admin\OneDrive - VENDIMAN PVT LTD\Amanda\DSR\vendekin\cashless 4 feb.csv'
df_cashless = pd.read_csv(csv_file_path_cashless)
df_cashless['Quantity'] = 1
df_cashless['PaymentStatus'] = 'SUCCESS'
df_cashless['Type'] = 'Cashless'
cashless = df_cashless.loc[:, ('Qrcode','Date', 'ProductName', 'Price', 'Location', 'Quantity','PaymentStatus', 'Type')]
cashless.head()
```

```python
#Refund

csv_file_path_refund =r'C:\Users\Admin\OneDrive - VENDIMAN PVT LTD\Amanda\DSR\vendekin\refund 4 feb.csv'
df_refund = pd.read_csv(csv_file_path_refund)
df_refund['Quantity'] = -1
df_refund['Type'] = 'Cashless'
df_refund['Actual_Type'] = 'Refund'
df_refund['Amount'] = -df_refund['Amount']
df_refund.rename(columns={'Amount': 'Price', 'RefundStatus': 'PaymentStatus' , 'RefundDate' : 'Date'}, inplace=True)
refund = df_refund.loc[:, ('Qrcode','Date', 'ProductName', 'Price', 'Location', 'Quantity','PaymentStatus', 'Type')]
refund.head()
```

```python
# Consolidating vendekin data - daily data

daily_data = pd.concat([cash,cashless,refund], ignore_index=True)
daily_data['DBSource'] = 'Vendiken'
daily_data.head()
```

```python
# Vendekin data - daily unique QR codes

unique_QRcode = pd.DataFrame(daily_data['Qrcode'].unique())
unique_QRcode = unique_QRcode.rename(columns={0: 'Qrcode'})
unique_QRcode

unique_QRcode.to_excel('feb3_QR codes.xlsx')
```

```python


# Vendekin data - DSR calculated table - vends calculations 

# cash vends cal
cash_transactions = daily_data[daily_data['Type'] == 'Cash']
cash_vends = cash_transactions.groupby(['Qrcode'])['Quantity'].sum().reset_index()
cash_vends_merged=pd.merge(unique_QRcode, cash_vends, on=['Qrcode'], how = 'outer')
cash_vends_merged.fillna(0, inplace = True)

# cashless vends cal

cashless_transactions = daily_data[daily_data['Type'] == 'Cashless']
cashless_vends = cashless_transactions.groupby(['Qrcode'])['Quantity'].sum().reset_index()
cashless_vends_merged=pd.merge(unique_QRcode, cashless_vends, on=['Qrcode'], how = 'outer')
cashless_vends_merged.fillna(0, inplace = True)

# # Total vends cal

total_vends = pd.merge(cash_vends_merged, cashless_vends_merged, on=['Qrcode'], how = 'outer' )
total_vends = total_vends.rename (columns = {'Quantity_x' : 'Cash Vends' , 'Quantity_y' : 'Cashless Vends' })
total_vends ['Total Vends'] = total_vends['Cash Vends'] + total_vends['Cashless Vends']





```

```python
# Vendekin data - DSR calculated table - amount calculations 


# cash amount cal
cash_transactions = daily_data[daily_data['Type'] == 'Cash']
cash_amount = cash_transactions.groupby(['Qrcode'])['Price'].sum().reset_index()
cash_amount_merged=pd.merge(unique_QRcode, cash_amount, on=['Qrcode'], how = 'outer')
cash_amount_merged.fillna(0, inplace = True)
cash_amount_merged

# cashless amount cal
cashless_transactions = daily_data[daily_data['Type'] == 'Cashless']
cashless_amount = cashless_transactions.groupby(['Qrcode'])['Price'].sum().reset_index()
cashless_amount_merged=pd.merge(unique_QRcode, cashless_amount, on=['Qrcode'], how = 'outer')
cashless_amount_merged.fillna(0, inplace = True)
cashless_amount_merged

# # # Total amount cal

total_amount = pd.merge(cash_amount_merged, cashless_amount_merged, on=['Qrcode'], how = 'outer' )
total_amount = total_amount.rename (columns = {'Price_x' : 'Cash Amount' , 'Price_y' : 'Cashless Amount' })
total_amount ['Total amount'] = total_amount['Cash Amount'] + total_amount['Cashless Amount']


vendekin_data = pd.merge(total_vends,total_amount , on = 'Qrcode' , how = 'outer')




```

```python
masterUID = pd.read_excel('UID QR master.xlsx')
vendekin = pd.merge(vendekin_data, masterUID, on ='Qrcode', how = 'inner')

vendekin = vendekin[['MachineUID', 'Cash Vends' , 'Cashless Vends' , 'Total Vends' , 'Cash Amount', 'Cashless Amount' , 'Total amount' ]]
vendekin['DB Source'] = 'Vendekin'

uvm = pd.read_excel('uvm23.xlsx')

uvm = uvm[uvm['Status'] == "SUCCESS"]
uvm = uvm[uvm['OperationLocationName'] != "Machine on lease"]
uvm['Type'] = uvm['PaymentMode'].apply(lambda x: 'Cash' if x == 'Cash' else 'Cashless')

unique_MachineUID = pd.DataFrame(uvm['MachineUID'].unique())
unique_MachineUID = unique_MachineUID.rename(columns={0: 'MachineUID'})
```

```python
# UVM data - DSR calculated table - vends calculations 

# uvm cash vends cal
uvm_cash_transactions = uvm[uvm['Type'] == 'Cash']
uvm_cash_vends = uvm_cash_transactions.groupby(['MachineUID'])['OrderId'].size().reset_index()
uvm_cash_vends_merged=pd.merge(unique_MachineUID, uvm_cash_vends, on=['MachineUID'], how = 'outer')
uvm_cash_vends_merged.fillna(0, inplace = True)


# # uvm cashless vends cal

uvm_cashless_transactions = uvm[uvm['Type'] == 'Cashless']
uvm_cashless_vends = uvm_cashless_transactions.groupby(['MachineUID'])['OrderId'].size().reset_index()
uvm_cashless_vends_merged=pd.merge(unique_MachineUID, uvm_cashless_vends, on=['MachineUID'], how = 'outer')
uvm_cashless_vends_merged.fillna(0, inplace = True)


# # # uvm Total vends cal

uvm_total_vends = pd.merge(uvm_cash_vends_merged, uvm_cashless_vends_merged, on=['MachineUID'], how = 'outer' )
uvm_total_vends = uvm_total_vends.rename (columns = {'OrderId_x' : 'Cash Vends' , 'OrderId_y' : 'Cashless Vends' })
uvm_total_vends ['Total Vends'] = uvm_total_vends['Cash Vends'] + uvm_total_vends['Cashless Vends']

uvm_total_vends.to_excel('uvm_vends.xlsx')
```

```python
# UVM data - DSR calculated table - amount calculations 


# cash amount cal
uvm_cash_transactions = uvm[uvm['Type'] == 'Cash']
uvm_cash_amount = uvm_cash_transactions.groupby(['MachineUID'])['Amount'].sum().reset_index()
uvm_cash_amount_merged=pd.merge(unique_MachineUID, uvm_cash_amount, on=['MachineUID'], how = 'outer')
uvm_cash_amount_merged.fillna(0, inplace = True)


# # cashless amount cal
uvm_cashless_transactions = uvm[uvm['Type'] == 'Cashless']
uvm_cashless_amount = uvm_cashless_transactions.groupby(['MachineUID'])['Amount'].sum().reset_index()
uvm_cashless_amount_merged=pd.merge(unique_MachineUID, uvm_cashless_amount, on=['MachineUID'], how = 'outer')
uvm_cashless_amount_merged.fillna(0, inplace = True)


# # # # Total amount cal

uvm_total_amount = pd.merge(uvm_cash_amount_merged, uvm_cashless_amount_merged, on=['MachineUID'], how = 'outer' )
uvm_total_amount = uvm_total_amount.rename (columns = {'Amount_x' : 'Cash Amount' , 'Amount_y' : 'Cashless Amount' })
uvm_total_amount ['Total amount'] = uvm_total_amount['Cash Amount'] + uvm_total_amount['Cashless Amount']

# uvm_total_amount.to_excel('filename.xlsx')
```
